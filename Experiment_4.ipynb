```python
import pandas as pd

data = pd.read_csv("processed_data/gender_PPN_removed_new_1.csv")
```

```python
print(data.info)
```

```python
male = data[data["Gender"] == "M"]
female = data[data["Gender"] == "F"]
```

```python
male_sents = male["Examples"].tolist()
male_sents_list = []
for l in male_sents:
    parsed_list = ast.literal_eval(l)
    for sent in parsed_list:
        if sent:
            male_sents_list.append(sent)

male_sents_set = set(male_sents_list)

print("Original:", len(male_sents_list))
print("Dedupped:", len(male_sents_set))
```

```python
female_sents = female["Examples"].tolist()
female_sents_list = []
for l in female_sents:
    parsed_list = ast.literal_eval(l)
    for sent in parsed_list:
        if sent:
            female_sents_list.append(sent)
            
female_sents_set = set(female_sents_list)

print("Original:", len(female_sents_list))
print("Dedupped:", len(female_sents_set))
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
 
def sentiment_scores(sentence):
 
    # Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()
    sentiment_dict = sid_obj.polarity_scores(sentence)
    
    return sentiment_dict['compound']
```

```python
male_scores = []

for sent in male_sents_set:
    male_scores.append(sentiment_scores(sent))
    
print("Male average:", sum(male_scores) / len(male_scores))

female_scores = []

for sent in female_sents_set:
    female_scores.append(sentiment_scores(sent))
    
print("Female average:",sum(female_scores) / len(female_scores))
```

```python
import scipy.stats as stats

results = stats.ttest_ind(female_scores, male_scores)

print(results)
```

```python
def print_senti(sentence):
    sid_obj = SentimentIntensityAnalyzer()
    sentiment_dict = sid_obj.polarity_scores(sentence)
    score = sentiment_dict['compound']
    print("{}; {}".format(score, sentence))
```

```python
for sent in female_sents_set:
    print_senti(sent)
```

```python
for sent in male_sents_set:
    print_senti(sent)
```
